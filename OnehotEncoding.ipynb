### What is OneHot Encoding
*   a method of converting categorical variables into a format that machine learning models can easily understand and use. Picture1.png






## How to implement



### Option 1: OneHotEncoder

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
df = pd.read_csv("https://raw.githubusercontent.com/ywen2021/CPSC392/main/Data/boutique.csv")
df
```

```python
categorical_features = [ "test_group","gender"]
onehot=OneHotEncoder(sparse_output=False)
one_hot_encoded = onehot.fit_transform(df[categorical_features])
# Convert to a DataFrame for easier interpretation
encoded_df = pd.DataFrame(one_hot_encoded, columns=onehot.get_feature_names_out([ "test_group","gender"]))

# Concatenate the encoded columns with the original DataFrame
df_encoded = pd.concat([df, encoded_df], axis=1)

df_encoded
```

```python
df_encoded=df_encoded.drop(columns=["test_group", "gender"])
df_encoded
```

## Option 2: Using get_dummies


```python
df = pd.read_csv("https://raw.githubusercontent.com/ywen2021/CPSC392/main/Data/boutique.csv")
df_encoded=pd.get_dummies(df,columns=[ "test_group","gender"],dtype=int)
df_encoded
```
